```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
results_path = "result.json"
```

```python
df_results = pd.read_json(results_path, orient="records")
```

```python
df_results[df_results["steps"] > 0].boxplot(column="expanded", by="method")
```

```python
df_results[(df_results["steps"] > 0) & (df_results["method"].isin(["Depth First", "Uniform Cost"]))].boxplot(column="expanded", by="method", figsize=(30,30), fontsize=40)
```

```python
df_results[(df_results["steps"] > 0) & (df_results["method"].isin(["Uniform Cost"]))].boxplot(column="expanded", by="method")
```

```python
df_results[(df_results["steps"] > 0) & (df_results["method"].isin(["Depth First", "Interactive Deepening Depth First"]))].boxplot(column="expanded", by="method")
```

```python
df_results.boxplot(column="expanded", by="method")
```

```python
for method in df_results["method"].unique():
#     fig = plt.figure() 
    plot = df_results[(df_results["steps"] > 0) & (df_results["method"] == method)].boxplot(column="expanded", by="method", figsize=(10,10))
    plt.title(f"{method} solveable results")
    plt.xlabel("Method")
    plt.ylabel("Steps")
    plt.savefig(f"results/{method}.png", formet="png", facecolor="white")
```
